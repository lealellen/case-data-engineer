```python
import pandas as pd
import glob

folders = ["data/bronze", "data/logs", "data/silver"]

for folder in folders:
    print(f"\n=== Lendo o primeiro arquivo em: {folder} ===")
    parquet_files = sorted(glob.glob(f"{folder}/part-*"))[:1]
    if not parquet_files:
        print("Nenhum arquivo encontrado!")
        continue
    for file in parquet_files:
        print(f"\nArquivo: {file}")
        try:
            df = pd.read_parquet(file)
            display(df.head())  
        except Exception as e:
            print(f"Erro ao ler {file}: {e}")
```
